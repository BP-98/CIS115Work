```python
#constant
#variable
inuse=True
prodName=['apple','banana','pear','mango','pumpkin']
prodQty=[12,5,6,2,4]
prodPriceDol=[0,0,1,1,5]
prodPriceCen=[75,40,00,10,50]
prodRestThresh=[5,3,3,5,2]
userAct=['List','Alter','Purchase','Leave']
alterOpt=['Add','Reduce','Remove','include']
revenueDol=0
revenueCen=0
#function
def itemList (array):
  aux=0
  for item in array:
    aux=aux+1
    print(aux,'for',item)
#code
#itemList(prodName) for menu
while inuse==True:
  aux=0
  for item in prodRestThresh:
    aux=aux+1
    if(prodRestThresh[aux-1]<prodQty[aux-1]):
      aux=aux
    else:
      print(prodName[aux-1]+'s needs a Restock')
  print('What would you like to do?')
  print('Current Revenue is $',revenueDol,'.',revenueCen)
  itemList(userAct)
  userSelect=int(input())
  if(userSelect==1):
    #List
    aux=0
    for item in prodName:
      aux=aux+1
      if (aux<=(len(prodName)-1)):
        print(prodQty[aux-1],item+'s, ')
      else:
        print(prodQty[aux-1],item+'s')
  elif(userSelect==2):
    #Alter
    itemList(alterOpt)
    userSelect=int(input())
    if(userSelect==1):
      #Add
      itemList(prodName)
      userSelect=int(input())
      userQty=int(input('How many to Add?'))
      if (userQty>0):
        prodQty[userSelect-1]=prodQty[userSelect-1]+userQty
    elif(userSelect==2):
      #Reduce
      itemList(prodName)
      userSelect=int(input())
      userQty=int(input('How many to Remove?'))
      if (userQty>0 and userQty<=prodQty[userSelect-1]):
        prodQty[userSelect-1]=prodQty[userSelect-1]-userQty
    elif(userSelect==3):
      itemList(prodName)
      userSelect=int(input())
      if(userSelect>0<len(prodName)+1):
        prodName.pop(userSelect-1)
        prodQty.pop(userSelect-1)
        prodPriceDol.pop(userSelect-1)
        prodPriceCen.pop(userSelect-1)
        prodRestThresh.pop(userSelect-1)
    elif(userSelect==4):
      userSelect=input('What Do you want to sell?: ')
      prodName.append(userSelect)
      userSelect=int(input('How many do you currently have in stock?: '))
      prodQty.append(userSelect)
      userSelect=int(input('How many Dollars Do they cost per?: '))
      prodPriceDol.append(userSelect)
      userSelect=int(input('How many Cents?: '))
      prodPriceCen.append(userSelect)
      userSelect=int(input('How mnay should we have until restock?: '))
      prodRestThresh.append(userSelect)
    else:
      print('Please Type a Valid Input')
  elif(userSelect==3):
    #Purchase
    aux=0
    for item in prodName:
      aux=aux+1
      print(aux,'for '+item+' at $',prodPriceDol[aux-1],'.',prodPriceCen[aux-1])
      userSelect=int(input('What would you like to buy?'))
      if(userSelect<=len(prodName) and userSelect>0):
        userQty=int(input('How many?'))
        if(userQty<=prodQty[userSelect-1] and userQty>0):
          prodQty[userSelect-1]=prodQty[userSelect-1]-userQty
          revenueDol=revenueDol+(userQty*prodPriceDol[userSelect-1])
          revenueCen=revenueCen+(userQty*prodPriceCen[userSelect-1])
        else:
          print('Unable to Purchase Quantity due to lack of Stock')
  elif(userSelect==4):
    #leave
    inuse=False
  else:
    print('Please Type a Valid Input')
print('GoodBye')
```
